````python
# Imports

import numpy as np
import pandas as pd
import statsmodels.api as sm

from IPython.display import display
from utils import read_cameras_dataset

%matplotlib inline

covariaveis = ['weight_lens', 'weight_body', 'focal_length_min']
````

````python
cameras_dataframe = read_cameras_dataset()
price_e_covariaveis_dataframe = cameras_dataframe[['price'] + covariaveis].dropna().astype(np.float64)

price_e_covariaveis_dataframe
````

Seja X a matriz das covariaveis e Y o vetor das respostas, temos:

````python
Y = price_e_covariaveis_dataframe['price']

vetor_beta0 = np.ones(len(price_e_covariaveis_dataframe))
X = np.vstack((vetor_beta0, price_e_covariaveis_dataframe[covariaveis].values.T)).T

display(pd.DataFrame(X))
display(Y)
````

Estimando $\hat{\beta} = (X^{T} X)^{-1} X^{T} Y$, temos:

````python
beta_estimado = np.linalg \
                  .inv(X.T.dot(X)) \
                  .dot(X.T) \
                  .dot(Y)

beta_estimado
````

Usando a biblioteca StatsModels, temos:

````python
fit = sm.OLS(Y, X).fit()
fit.summary()
````
