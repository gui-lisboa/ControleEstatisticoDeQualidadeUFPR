## Seção II.7

### Questão 1

Suponha que $|A|=0$. É possível que todos os valores próprios de A sejam positivos?

Não, pois dado que $|A| = 0$, temos $|A| = |Λ|= \prod_{j=1}^{p}λ_j = 0$. Para que o produtório seja igual a 0, pelo menos um autovalor deve ser igual a 0, logo, não positivo.

### Questão 2

Suponha que todos os valores próprios de alguma matriz (quadrada) A sejam diferentes de zero. A inversa $A^{−1}$ existe?

Para que a matrix quadrada admita inversa, $|A| \neq 0$. Como todos os autovalores são diferentes de 0, temos que $\prod_{j=1}^{p}λ_j \neq 0$ e assim, $\prod_{j=1}^{p}λ_j = |Λ| = |A| \neq 0$, logo, a inversa $A^{-1}$ existe.

### Questão 16

Mostre que $posto(H)=tr(H)=n−1$.

Seja $H = I_n - \frac{1}{n} 1_n 1_n^T$, temos que $1_n 1_n^T = n$,

Reescrevendo $H$, temos: $H = I_n - [\frac{n}{n}$]

Como $\frac{n}{n} = 1$, temos $H = I_n - [1]$

Como $tr(A + B) = tr(A) + tr(B)$, temos: $tr(H) = tr(I_n) - tr([1])$

Como o traço é a soma da diagonal, para uma matriz identidade com $n$ linhas o traço é igual a $n$, logo:

$tr(H) = n - 1$

Encontre os momentos condicionais $E(X_2|x_1)$ e $E(X_1|x_2)$ para a função de densidade no Exemplo IV.5.

$$
E(X_2|x_1) = \int_{0}^{1} x_2 f(x_2|x_1) dx_2 \\
E(X_2|x_1) = \int_{0}^{1} x_2 \frac{\frac{1}{2}x_1 + \frac{3}{2}x_2}{\frac{1}{2}x_1 + \frac{3}{4}} dx_2 \\
E(X_2|x_1) = \frac{x_1}{3 + 2x_1} + \frac{2}{3 + 2x_1} = \frac{x_1 + 2}{3 + 2x_1}

$$
E(X_1|x_2) = \int_{0}^{1} x_1 f(x_1|x_2) dx_1 \\

### Questão 11

Considere a função de densidade $f(x_1,x_2) = \frac{1}{8x_2} exp(−(\frac{x_1}{2x_2} + \frac{x_2}{4}))$, $x1,x2 \gt 0$.
Encontre $f(x_2)$ e $f(x_1|x_2)$. Encontre também a melhor aproximação de $X_1$ por uma função de $X_2$.

$f_{X_2}(x_2)$:

$$
f_{X_2}(x_2) = \int_{0}^{\infty} f(x_1, x_2) dx_1 \\
f_{X_2}(x_2) = \int_{0}^{\infty} \frac{1}{8x_2} exp(−(\frac{x_1}{2x_2} + \frac{x_2}{4}) dx_1 \\
f_{X_2}(x_2) = \frac{1}{4} exp(-\frac{x_2}{4})
$$

$f_(x_1|x_2)$

$$
f_(x_1|x_2) = \frac{f(x_1, x_2)}{f_{X_2}(x_2)} \\
f_(x_1|x_2) = \frac{\frac{1}{8x_2} exp(−(\frac{x_1}{2x_2} + \frac{x_2}{4}))}{\frac{1}{4} exp(-\frac{x_2}{4})} \\
f_(x_1|x_2) = \frac{1}{2x_2 exp(\frac{x_1}{2x_2})}
$$

Aproximação de $X_1$ por $X_2$

Seja $E(X_1 | X_2)$ a melhor aproximação de $X_1$ por $f(x_2) = \frac{1}{4} exp(-\frac{x_2}{4})$ de $X_2$ (Exponencial($\lambda = \frac{1}{4}$) ), temos:

$$
E(X_1) = E(E(X_1 | X_2)) = E(X_2) = \frac{1}{\lambda} = 4
$$

Aplicar as técnicas fatoriais ao conjunto de dados: Time Budget Data ou dados temporais de orçamento em In Volle (1985). Nestes arquivo de dados, 28 indivíduos identificados de acordo com o sexo, país onde vivem, atividade profissional e estado civil, indicando o tempo que cada pessoa gastou em 10 categorias de atividades ao longo de 100 dias, 100×24h = 2400 horas total em cada linha, em 1976.

````python
import pandas as pd
import numpy as np
import seaborn as sns

sns.set(rc={'figure.figsize':(15,15)})

dados = pd.read_table("./dados/time_budget_data.txt", sep=" ", index_col=0)

matriz_correlação = dados.corr()
heatmap = sns.heatmap(matriz_correlação, vmin=-1, vmax=1, annot=True)
````

````python
n = len(dados.index)
m = len(dados.columns)

H = np.identity(n) - np.full(shape=(n,n), fill_value= 1/n)
D = np.diag(1/np.sqrt(((n-1)/n) * np.diag(dados.cov().values)))

X1 = H.dot(dados).dot(D)/np.sqrt(n)

auto_valores, auto_vetores = np.linalg.eig(X1.T.dot(X1))

# arruma para fator próximo de 0 que assume valor negativo
auto_valores = auto_valores.round(10)

auto_valores
````

````python
soma_auto_valores = auto_valores.sum()
dominância = auto_valores.cumsum()/soma_auto_valores

dominância = pd.DataFrame({
    "fator": dados.columns,
    "dominancia": dominância,
})

dominância
````

````python
W = np.sqrt(auto_valores) * (auto_vetores.T)

W = pd.DataFrame(W, columns=dados.columns, index=dados.columns)

W
````

````python
sns.set(rc={'figure.figsize':(8,8)})
scatter = sns.scatterplot(x="prof", y="tran", s=200, data=W.T)
scatter.axvline()
scatter.axhline()

for i in range(W.iloc[:2].columns.size):
    scatter.annotate(W.iloc[:2, i].name, W.iloc[:2, i].values + .01)

W.iloc[:2]
````

````python
Z = X1.dot(auto_vetores)

Z = pd.DataFrame(Z, columns=dados.columns, index=dados.index)

Z.head()
````
