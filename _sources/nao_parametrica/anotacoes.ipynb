## Testes para uma amostra

### Teste Binomial

O foco principal é problemas com dados dicotômicos que consistem em n repetições de ensaios de Bernoulli independetes, todas as observações são provenientes de uma mesma população e a probabilidade p de sucesso não se altera com a repetição do experimento.

### Estatística de teste

$$
Y = T \text{, onde T corresponde o número de sucessos} \\
Y^* = \frac{B - E_{p_0}(Y)}{\sqrt{Var_{p0}(Y)}} = \frac{B - np_0}{\sqrt{np_0(1-p_0)}} =  T^* \sim  N(0,1) \\
\text{correção de continuidade: } Y^* = \frac{(Y \pm 0,5) - E_{p_0}(Y)}{\sqrt{Var_{p0}(Y)}}
$$

### Teste do Quantil

Nível da mensuração é pelo menos ordinal, utiliza o teste da Binomial.

$p^*$, o p-ésimo quantil; $x^*$ um valor especificado;
$T_1$ o número de observações menores ou iguais a $x^*$; $T_2$ o número de observações menores que $x^*$; $T_1 = T_2$ se nenhuma observação for igual a $x^*$; caso contrário, $T_1 > T_2$.

$$
P(Y \le q_1) = \alpha_1 \text{ e } \\
P(Y \gt q_2) = \alpha_2 \text{ ou } P(Y \le q_2) = 1 - \alpha_2 \\
$$

Rejeitar $H_0$ se $T_1$ é menor ou igual a $q_1$ ou se $T_2$ é maior a $q_2$ ao nível de significância $\alpha$.

## Testes de aderência

### Teste Qui-Quadrado de Aderência

Comparar as frequências das categorias da variável de interesse obtidas na amostra com as frequências esperadas baseadas em um distribuição de probabilidade específica.

Escala nominal.

$$
 Q = \sum^{k}_{i=1} \frac{(O_i - E_i)^2}{E_i} \\
$$

em que:

$k$ = Número de categorias

$O_i$ = Frequência observada na categoria $i$

$E_i$ = Frequência esperada na categoria $i$


### Teste de Kolmogorov-Smirnov de Aderência

Nível de mensuração deve seguir ao menos uma escala ordinal (mais comum intervalar ou de razão). Exige que a distribuição a ser testada seja plenamente especificada. Aplicável a pequenas amostras.
A estatística T de Kolmogorov-Smirnov, que corresponde a maior distância vertical entre $F^*(x)$ e $S(x)$:

$$
T = supremo|F^*(x) - S(x)|
$$

Caso o teste seja unilateral, usar $T^+$ (unilateral à direita) ou $T^-$ (unilateral à esquerda).

$$
T^+ = supremo(F^*(x) - S(x)) \text{ maior distância positiva} \\
T^- = supremo(S(x) - F^*(x)) \text{ maior distância negativa}
$$

O teste tende a aceitar a hipótese $H_0$ mais vezes do que deveria.

### Teste Lilliefors para distribuição normal

Teste KS com modificação introduzida por Lilliefors. Verifica se os dados amostrais são provenientes de uma distribuição normal, com média e desvio-padrão não especificados.

Média amostral e desvio-padrão amostral:

$$
    \bar{X} = \frac{1}{n} \sum^n_{i=1}X_i \\
    S = \sqrt{\frac{1}{n-1} \sum^n_{i=1} (X_i - \bar{X})^2}
$$

Escores padronizados $Z_i$:

$$
Z_i = \frac{X_i - \bar{X}}{S} \text{ , i= 1, 2, ... n}
$$

Com escores ordenados da amostrar, obter $S(x)$

### Teste de Shapiro-Wilk

Só é aplicável para verificar normalidade. A escala de mensuração é ao menos ordinal. Capaz de detectar desvios da normalidade devido à assimetria e curtose. Compara a variância amostral clássica com a variância estimada via dsitribuição amostral das estatísticas de teste.

$$
W = \frac{[\sum^k_{i=1} a_i (x_{(n+1-i)} - x_{(i)})]^2}{\sum^n_{i=1}(x_i - \bar{x})^2}
$$

sendo

$$
\begin{equation}
    k= \begin{cases}
        \frac{(n+1)}{2} \text{ se n impar} \\
        \frac{(n)}{2} \text{ se n par} \\
    \end{cases}
\end{equation}
$$

W próximo a 1 indicam normalidade
